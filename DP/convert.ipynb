```python
import numpy as np
import rasterio
from rasterio.transform import from_origin

# načítanie npz
data = np.load("zdroj/clippedMaskMounds.npz")

print(data.files)  # pozri aké premenné sú vnútri

array = data[data.files[0]]  # väčšinou je tam jeden array

transform = from_origin(0, 0, 1, 1)

# uloženie do tif
with rasterio.open(
    "zdroj/clippedMask.tif",
    "w",
    driver="GTiff",
    height=array.shape[0],
    width=array.shape[1],
    count=1,
    dtype=array.dtype,
    transform=transform
) as dst:
    dst.write(array, 1)
```
